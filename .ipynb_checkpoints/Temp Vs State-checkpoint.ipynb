```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as numpy
```

```python
#import csv files and #turn csv into a dataframe
data = pd.read_csv('GlobalLandTemperaturesByState.csv')
data.head()
```

```python
#remove null values
data1=data.dropna()
data1.count()
```

```python
#Extracting year and month from the data(data1) 
data1['Year'] = data1['dt'].str[:4]
data1['Month'] = data1['dt'].str[6:7]
data1.head()
```

```python
#again extracting june july august data from  data1 and name the file (june_july_august_data)
june_july_august_data=data1[(data1['Month'] >= "6") & (data1['Month'] <= "8")]
june_july_august_data.head()
```

# JUNE JULY AUGUST DATA OF NEW_YORK WITH TEMPERATURE

```python
# From (june_july_august_data) Extracting only New_York 
New_York_data=june_july_august_data[(june_july_august_data['State'] == 'New York')]
New_York_data.head()
```

```python
#Taking out mean  of Average temperature  for  each  same year 
New_York_State_group = New_York_data.groupby(["State","Year"], as_index=False)
New_York_State_temp_mean = New_York_State_group.mean()
```

```python
New_York_State_temp_mean.head()
```

```python
#Plotting  New York VS  Average Temperature
New_York_State_temp_mean.plot("Year",'AverageTemperature',figsize=(15,8))
plt.ylabel("AverageTemperature", fontsize=12)
plt.xlabel("Year")
plt.title("New York VS Average Temperature", fontsize=18)
plt.show
```

# WORKING ON EMISSION FILE

```python
#import csv files and #turn csv into a dataframe
df = pd.read_csv('CO_level_2000_-.csv')
df.head()
```

```python
#Taking out the columns 
df1=df[['State', 'Date','Arithmetic Mean']]
df1.head()
```

```python
#Extracting year and month data
df1['Year'] = df1['Date'].str[:4]
df1['Month'] = df1['Date'].str[6:7]
df1.head()
```

```python
# extracting june july august data and name it (JJA)
JJA_data=df1[(df1['Month'] >= "6") & (df1['Month'] <= "8")]
JJA_data.head()
```

```python
JJA_data=JJA_data.rename(columns={"Arithmetic Mean": "Emission Mean"})
```

```python
JJA_data.head()
```

# Nasa_data (Global_Mean_Estimates_based_on_Land_and_Ocean_Data)

```python
Nasa_data = pd.read_csv('Nasa_Data.csv')
Nasa_data.head()
```

```python
#Plotting  Year VS  No_Smoothing
#Plotting  Year VS  Lowess(5)
Nasa_data.plot("Year",'No_Smoothing',figsize=(15,8))
Nasa_data.plot("Year",'Lowess 5',figsize=(15,8))
plt.ylabel("No_Smoothing", fontsize=12)
plt.ylabel("Lowess 5", fontsize=12)
plt.xlabel("Year")
plt.title("plot of  with respect to Emission Mean", fontsize=18)
plt.show
```

```python
# Combine Plot of  Year VS  (No_Smoothing and Lowess(5)
ax = plt.gca()
Nasa_data.plot(kind='line',x='Year',y='No_Smoothing',ax=ax,figsize=(15,8))
Nasa_data.plot(kind='line',x='Year',y='Lowess 5', color='red',ax=ax,figsize=(15,8))
plt.title("Combined Plot of  No_Smoothing and Lowess (5)with respect to Year", fontsize=18)

plt.show()
```
