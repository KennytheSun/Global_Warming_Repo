```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as numpy
```

```python
#import csv files and #turn csv into a dataframe
csv = "../Rescources/GlobalLandTemperaturesByState.csv"
data = pd.read_csv(csv)
data.head()
```

```python
#import csv files and #turn csv into a dataframe
csv2 = "../Rescources/CO_level_2000_-.zip"
df = pd.read_csv(csv2)
df.head()
```

```python
#remove null values
data1=data.dropna()
data1.count()
```

```python
data1.head()
```

```python
#Extracting year and month from the data(data1) 
data1['Year'] = data1['dt'].str[:4]
data1['Month'] = data1['dt'].str[5:7]
data1.head()
```

```python
#again extracting DEC, JAN , FEB  data from  data1 and name the file (winter_data)
winter_data=data1[(data1['Month'] == "12") & (data1['Month'] <= "2")]
winter_data.head()
```

```python
winter_data['Year'] = winter_data['Year'].astype(float)
```

```python
winter_data.head()
```

```python
# taking out 2000 to 2012 data from data1 and save into year_data
winter_data=winter_data[(winter_data['Year'] >= 2000.0) & (winter_data['Year'] <= 2012.0)]
winter_data.head()
```

```python
#Taking out the columns 
df1=df[['State', 'Date','Arithmetic Mean']]
df1.head()
```

```python
#Extracting year and month data
df1['Year'] = df1['Date'].str[:4]
df1['Month'] = df1['Date'].str[5:7]
df1.tail()
```

```python
# extracting dec,jan, feb data and name it (Ewinter)
Ewinter_data = df1[(df1['Month'] == "01") | (df1['Month'] == "02") | (df1['Month'] == "12")]
Ewinter_data.head()
```

```python
Ewinter_data=Ewinter_data.rename(columns={"Arithmetic Mean": "Emission Mean"})
```

```python
Ewinter_data=Ewinter_data[(Ewinter_data['Year'] >= "2000") & (Ewinter_data['Year'] <= "2012")]
Ewinter_data.head()
```

## WINTER DATA OF CALIFORNIA  WITH TEMPERATURE AND EMISSION

```python

# From (winter_data) Extracting only California
#Taking out mean  of Average temperature  for  each  same year
#Plotting  California VS  Average Temperature
California_data=winter_data[(winter_data['State'] == 'California')]
California_State_group = California_data.groupby(["State","Year"], as_index=False)
California_State_temp_mean = California_State_group.mean()

# From (Ewinter_data) Extracting only California
#Taking out mean  of Average temperature  for  each  same year
#Plotting  California VS  Average Temperature
California_data=Ewinter_data[(Ewinter_data['State'] == 'California')]
California_State_group = California_data.groupby(["State","Year"], as_index=False)
California_State_Emission_mean = California_State_group.mean()




# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(California_State_temp_mean.Year, California_State_temp_mean["AverageTemperature"], color="red")
# set x-axis label
ax.set_xlabel("Year",fontsize=14)
# set y-axis label
ax.set_ylabel("AverageTemperature",color="red",fontsize=14)

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(California_State_temp_mean.Year, California_State_Emission_mean["Emission Mean"] ,color="blue")
ax2.set_ylabel("Emission Mean",color="blue",fontsize=14)
plt.title("Emission Mean and Temperature plot of California (winter)", fontsize=12)
plt.show()

```
