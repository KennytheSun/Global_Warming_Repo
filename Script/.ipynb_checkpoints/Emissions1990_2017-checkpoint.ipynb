```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Global_emissions_file = "../Resources/StateEmissions1990.csv"
Global_temp_file = "../Resources/GlobalTemperatures.csv"

emission_df = pd.read_csv(Global_emissions_file)
temp_df = pd.read_csv(Global_temp_file)

emission_df = emission_df.rename(columns={"Million metric tons of carbon dioxide" : "State"})
emission_df = emission_df.set_index('State')
emission_df = emission_df.drop(columns=['Percent', 'Absolute', 'Unnamed: 31'] )
emission_df
```

```python
x_axis = []
avg_emissions_year = []
start_year = 1990
end_year = 2017
current_year = 1990

for i in range (start_year, end_year + 1):
    x_axis.append(current_year)
    avg_emissions_year.append(emission_df[f"{current_year}"].sum())
    current_year += 1
    
```

```python
plt.plot(x_axis, avg_emissions_year)
plt.xlabel('Year')
plt.ylabel('Million Metric tons of CO2')
plt.title('Total number of CO2 Emissions in the U.S. per year')
```

```python
state = input("What state would you like to see?: ")
state = state.lower()
state = state.capitalize()

state_emissions = emission_df.loc[f"{state}"]
plt.plot(x_axis, state_emissions)
plt.xlabel('Year')
plt.ylabel('Million Metric Tons of C02')
plt.title(f'C02 emissions of {state} per year')
```

```python
percent_increase = (1 - (emission_df["1990"].sum() + emission_df["1991"].sum() + emission_df["1992"].sum())/
                         (emission_df["2015"].sum() + emission_df["2016"].sum() + emission_df["2017"].sum() )) * 100

print (f"{percent_increase.round(2)}% increase") 
```
